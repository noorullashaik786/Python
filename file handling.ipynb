```python
from functools import reduce
def product(x,y):
  return x*y
ans = reduce (product,[1,2,3,4])
print(ans)
```

```python
import functools
import operator
print(functools.reduce (operator.add,(1,2)))
print(functools.reduce (operator.concat,("python ","iot")))
```

```python
from itertools import accumulate
a = [1,2,3,4]
print(list(accumulate(a,lambda x,y:x+y)))
```

```python
a = [-1,-2,-3,1,2,3]
result = list(filter(lambda x:x>=0,a))
print(result)
```
