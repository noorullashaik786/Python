9.Write a program to check whether a number is prime or not using while loop.


```python

num = int(input("Enter a number ( greater than 1)"))
f = 0
i = 2
while i <= num / 2:
    if num % i == 0:
        f=1
        break
    i=i+1

if f==0:
    print("The entered number is a PRIME number")
else:
    print("The entered number is not a PRIME number")
```

11.Write a program to find the sum of the digits of a number accepted from the user.

```python
n=int(input("Enter number"))

sum=0

for digit in str(n):
    sum=sum+int(digit)

print("Sum of digits",sum)
```

15. Write a program to print the Fibonacci series till n terms (Accept n from user) using while loop.

```python
nterms = int(input("How many terms? "))
n1, n2 = 0, 1
count = 0
if nterms <= 0:
   print("Please enter a positive integer")
elif nterms == 1:
   print("Fibonacci sequence upto",nterms,":")
   print(n1)
else:
   print("Fibonacci sequence:")
   while count < nterms:
       print(n1)
       nth = n1 + n2
       n1 = n2
       n2 = nth
       count += 1
```

16. Write a program to print the factorial of a number accepted from user.

```python
num = int(input("enter a number "))
fac = 1
for i in range(1,num+1):
  fac = fac * i
  print(fac)
```
