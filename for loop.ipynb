```python
bill = int(input("enter the no of units :"))
if bill >= 1 and  bill <= 100 :
  amt = bill*10
  print("amount is :",amt)
elif bill >= 100 and bill <= 200 :
  amt = bill*15
  print("amount is :",amt)
elif bill >= 200 and bill <= 300 :
  amt = bill*20
  print("amount is :",amt)
else :
  amt = bill*25
  print("amount is :",amt)
```
