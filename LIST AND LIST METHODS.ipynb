UPDATING

```python
lst = ["mango",1,2,3]
lst[1]=10
print(lst)     #parent variable value also changes
```

```python
#APPEND
prime = [2,3,5,7,9,11]
prime.append(13)
print(prime)     #add a new value in last of list
list = [1,3,5,7,9]
list.append([4,5,8])
print(list)
list[5][2]       # 2d array
```

```python
#INSERT
even = [2,4,6,10,12]
even.insert(3,8)
print(even)
```

```python
#EXTEND
list = [1,2,3,4,5]
list.extend([6,7,8])
print(list)
list.extend([[9,10,[11,12,13]]])      #3d array
print(list)
list[8][2][2]
```

DELETING

```python
lst =[1,2,3,4,5,10]
del lst[-1]
print(lst)
```

```python
#POP
lst =[1,2,3,4,5,10]
lst.pop()             # delete last element by default
print(lst)
```

```python
#POP
lst =[1,2,3,4,5,10]
lst.pop(3)           #(position)
print(lst)
```

```python
#REMOVE
lst =[1,2,3,4,5]
lst.remove(4)          #(element)
print(lst)
```

LIST METHODS

```python
#1.COPY
l = [1,3,5,7,9]
l2 = l.copy()
print(l2)
```

```python
#2.index
lst = [1,2,3,4]
lst2 = lst.index(3)           #(element)   returns position
print(lst2)
```

```python
#3.COUNT
list1 = [1,2,3,3,4,5,4,3,2,1]
lst= list1.count(3)
print(lst)
```

```python
#4.REVERSE
list1 = [1,2,3,4,5]
list1.reverse()
print(list1)
list2 = list1.reverse()            #variable not assigned it retuns none
print(list2)
```

```python
#5.SORT
lst = [7,4,9,2,6,1,9]
lst.sort()                     #assending order
print(lst)
ls = lst.sort()                  #variable not assigned it retuns none
print(ls)
lst.sort(reverse=True)           #descending order
print(lst)
```

```python
#MAXIMUM AND MINIMUM
lst = [1,3,-5,9,10]
max(lst)
min(lst)
var = "AXBZYC"
max(var)
min(var)
var2 = "AazZ"
max(var2)
min(var)
```
