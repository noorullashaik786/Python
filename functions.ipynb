```python
def odd_even(num):
  if num%2==0:
    return("even")
  else :
    return("odd")
result = odd_even(10)
print(result)
```

```python
def positive_negative(num):
  if num > 0 :
    return("positive")
  else :
    return("negative")
result= positive_negative(-4)
print(result)
```

```python
def start_stop(n1,n2):
  empty = []
  for i in range(n1,n2+1):
      empty.append(i)
  return(empty)
n1 = int(input("enter start number:"))
n2 = int(input('enter ending number:'))
result= start_stop(n1,n2)
print(result)
```

```python
def palindrome():
  if st==rev:
    return("palindrome")
  else :
    return("not palindrome")
st = "mam"
rev = st[::-1]
result = palindrome()
print(result)
```

ARGUEMENTS

```python
#positional arguements
def func(lan1,lan2):
  print(lan1,lan2)
func("python","java")
```

```python
#keyword arguements
def func(lan1,lan2):
  print(lan1,lan2)
func(lan2="python",lan1="java")
```

```python
#default arguments
def func(lan1,lan2,lan3="iot"):
  print(lan1,lan2,lan3)
func("python","java")
def func(lan1,lan2,lan3="iot"):#lan4):             #after default no another arguements
  print(lan1,lan2,lan3)#lan4)
func("python","java")#'html')
```

```python
#variable length arguement
def func(*lan1):
  print(lan1)                              #  * args
func('python ','java','html')

def func(**lan):
  print(lan)                       #  * kwargs  (prints in dictionary)
func(lan1='python ',lan2='java',lan3='html')
```

```python
#local variable
def fu():
  a = 10
  b = 20
  print(a,b)
fu()
#print(a,b)          #error because local variable
```

```python
#global variable
A = 5
def fu():
  a = 10+A
  b = 20+A
  print(a,b,A)
fu()
```

```python
a = 5
def func():
  a = 10
  b = 20
  print(globals()['a'],a,b)
func()
print(a)
```

```python
#RECURSION FUNCTION
def nxt_palindrome(num):
  rev =str(num)[::-1]
  if rev == str(num):
    print(f"{num}is a palindrome")
  else:
    nxt_palindrome(num+1)
num =int(input("enter a number:"))
nxt_palindrome(num)
```

```python
def palindrome():
  n = 122
  temp = n
  rev = 0
  while n > 0:
    rem = n%10
    rev = (rev*10)+rem
    n = n//10
  if temp == rev:
    print("palindrome")
  else :
    print("not a palindrome")
palindrome()
```
