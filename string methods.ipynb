PASSWORD VALIDATION

```python
password = input("enter the password : ")
upper_count = 0
lower_count = 0
special_count = 0
number_count = 0
for i in password :
  if i.isupper():
    upper_count+=1
  elif i.islower():
    lower_count+=1
  elif i.isdigit():
    number_count+=1
  else :
    special_count+=1
if len(password)>=8 and upper_count>=2 and lower_count>=1 and special_count>=1 and number_count>=2 :
  print("VALID PASSWORD")
else :
  print("INVALID PASSWORD")

```
